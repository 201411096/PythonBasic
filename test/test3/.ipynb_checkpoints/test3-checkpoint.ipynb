```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

from scipy import stats
from matplotlib import font_manager, rc
from statsmodels.sandbox.regression.predstd import wls_prediction_std
from sklearn.datasets import make_regression

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
X = [24,25,26,27,28,29,30,31,35] # 온도 X
Y = [22,25,28,30,33,34,36,32,38] # 아이스크림 판매량 Y

slope, intercept, r_value, p_value, stderr = stats.linregress(X,Y) # intercept는 y절편

print('분석 결과 :==============================')
print('r_value(상관계수) : ' + str(r_value))
if abs(p_value) <0.05:
    print('유의미한 결과')
else:
    print('유의미하지 않은 결과')

# r_value ?
plt.scatter(X, Y)
plt.grid()
plt.xlabel('기온')
plt.ylabel('아이스크림 판매량')
X1 = np.array(X)
plt.plot(X1, X1*slope+intercept, c='red')
```

```python
data2 = pd.read_csv('dairy.csv')
data2

# 판매량=소비량???
temp = data2[ (data2['dairy']=='치즈') & (data2['year']>=2000)] # 치즈이면서 2000년 이후
temp2 = temp[['year', 'consumption']].groupby('year').sum() # 연도별로 소비량을 그룹화

plt.figure(figsize=(16,5))
plt.xticks(temp['year'])

plt.plot(temp2['consumption'])
plt.xlabel('year')
plt.title('치즈 판매량')
plt.grid()
```
