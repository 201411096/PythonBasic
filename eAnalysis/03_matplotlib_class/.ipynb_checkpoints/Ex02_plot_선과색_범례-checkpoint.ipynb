```python
%matplotlib inline
    
import matplotlib.pyplot as plt
import numpy as np

""" 0. 플롯 스타일 """
#plt.style.use('classic')
plt.style.use('seaborn-whitegrid')

# np.linspace
# 선형 간격의 벡터 생성
# linspace(x1, x2, n) x1부터 x2까지 n개의 점을 생성

x = np.linspace(0, 10, 100) # 0부터 10까지 100개를 뽑음
plt.plot(np.sin(x))
```

```python
""" 1. 색상지정 """
plt.plot(np.sin(x-0), color='red')     # 색상 이름 지정
plt.plot(np.sin(x-1), color='b')       # 색상코드명
plt.plot(np.sin(x-2), color='#FFDD44') # 16진수 RGB
plt.plot(np.sin(x-3), color='chartreuse') # html 색상 이름
plt.plot(np.sin(x-4), color='0.75') # 0-1 사이 값으로 회색
plt.plot(np.sin(x-5), color=(1.0, 0.5, 0.2)) # rgb를 튜플 형식으로
```

```python
""" 2. 선스타일 """

# plt.plot(np.sin(x-0), linestyle='solid');
# plt.plot(np.sin(x-1), linestyle='dashed');
# plt.plot(np.sin(x-2), linestyle='dotted');
# plt.plot(np.sin(x-3), linestyle='dashdot');

plt.plot(np.sin(x-0), linestyle='-');
plt.plot(np.sin(x-1), linestyle='--');
plt.plot(np.sin(x-2), linestyle=':');
plt.plot(np.sin(x-3), linestyle='-.');
```

## r(red) g(green) b(blue) c(cyan) m(magenta) y(yellow) k(black) 

```python
""" 3. 선과 스타일을 간결하게 (대략 읽는 수준으로 ) """

plt.plot(np.sin(x-0), 'g--');
plt.plot(np.sin(x-1), ':m');
plt.plot(np.sin(x-2), 'k-.'); # dashdot + black
plt.plot(np.sin(x-3), 'r:'); # red + dotted
```

```python
''' 4. 범례 '''

plt.plot(np.sin(x-0), 'g--', label='first');
plt.plot(np.sin(x-1), ':m', label='second');
plt.plot(np.sin(x-2), 'k-.', label='third'); # dashdot + black
plt.plot(np.sin(x-3), 'r:', label='fourth'); # red + dotted

# 범례 위치 : 오른쪽 상단 (1), 왼쪽 상단(2)
plt.legend()
```
