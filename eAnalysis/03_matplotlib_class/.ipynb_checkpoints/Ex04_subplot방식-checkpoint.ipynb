```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(그림틀=공간?)을 만들고 거기에 subplot을 추가 하는 방식 """
fig = plt.figure(figsize= (16, 5))
ax1 = fig.add_subplot(1,2,1) # 1행 2열의 첫번째
ax2 = fig.add_subplot(1,2,2)
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(100).cumsum())
```

```python
""" 2. figure로 공간을 만들고 파이플롯(pyplot)의 subplot()으로 각각의 패널 지정 """

fig = plt.figure(figsize= (16, 5))
ax1 = plt.subplot(1,2,1) # 1행 2열의 첫번째
ax2 = plt.subplot(1,2,2)
ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(100).cumsum())
```

```python
""" 3. 간단하게 ax 객체를 이용하기 """

fig, ax = plt.subplots(1,2) # return 값이 두개임 (figure axes)
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```

```python
''' 4. 그래프 전체 크기를 지정 '''

fig, ax = plt.subplots(1,2, figsize=(17,5)) # return 값이 두개임 (figure axes) -> axes가 subplot인듯
ax[0].plot(np.random.randn(100))
ax[1].plot(np.random.randn(100).cumsum())
```

## 서브플롯 개념

```python
plt.figure(figsize=(12,6))

plt.subplot(221) # 2행 2열의 첫번째 영역 # 전체영역을 2행 2열로 나누었을때 첫번째 영역
plt.subplot(222) # 2행 2열의 두번째 영역 # 전체영역을 2행 2열로 나누었을때 두번째 영역
plt.subplot(212) # 2행 1열의 2번째 영역 # 전체영역을 2행 1열로 나누었을때 두번째 영역

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(211)
plt.subplot(223)
plt.subplot(224)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.subplot(2, 1, 1)
plt.subplot(2, 3, (4,5))
plt.subplot(2, 3, 6)

plt.show()
```
