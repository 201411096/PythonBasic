## 하나의 화면에 여러 개의 그래프 그려보기

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 그래프가 겹쳐짐
# plt.plot(np.random.randn(100))
# plt.plot(np.random.randn(100).cumsum())

fig = plt.figure(figsize= (16, 5))
ax1 = fig.add_subplot(1,2,1) # 1행 2열의 첫번째
ax2 = fig.add_subplot(1,2,2)

ax1.plot(np.random.randn(100))
ax2.plot(np.random.randn(100).cumsum())
```

```python
# x, y 축 값 범위 지정 

fig = plt.figure()
ax = fig.add_subplot(111) # (1,1,1) 동일
ax.plot(np.random.randn(50), 'b+--')
ax.set_xlim([1, 100]) # x축의 범위를 변경 (x limit?..)
ax.set_ylim([-5, 5])
ax.set_xlabel('data')
ax.set_ylabel('temp')
ax.set_title('My Graph')
```

## 그래프 속성 지정

- title() : 
- xlabel() : 
- ylabel() : 
- grid() : 격자 눈금
- text() : 글씨를 출력 (tooltip을 간단하게 하고자 할 때)

```python
# plt에서 그래프 속성 지정

plt.plot(np.random.randn(10))
plt.xlabel('data')
plt.ylabel('temp')
plt.text(3, 2, 'random value')
plt.grid()
```

### x 축 간격

```python
# figure로 전체 그래프의 액자 같은 개념으로 액자 크기 조절이 가능하다
fig = plt.figure()
ax = fig.add_subplot(111) # (1,1,1) 동일
ax.plot(np.random.randn(50), 'b+--')
ax.set_xticks([10, 20, 30, 40, 50, 60])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five', 'six'])
```
