# 고속버스 데이타셋에서 인사이트를 도출

[예] 

1. 우등고속에 대한 데이타 필터링(추출)

2. 경부선에서 총이용인원이 1000명 이상인 데이타 필터링(추출)

3. 그 외 데이타를 찾아내시오

```python
import pandas as pd

df = pd.read_excel('data/고속버스.xlsx') 
       # sheetname='highway'  encoding='cp949' 추가 안해도 실행  
#df.head()  # 상위 5개만 보임
df
```

[연습]

```python
# 1. 우등고속에 대한 데이타 필터링(추출)

#df.groupby('차종').count()
#df.groupby('차종').mean()
df.groupby('차종').count().loc['우등']
#df[df['차종']=='우등'].count()
```

```python
# 2. 경부선에서 총이용인원이 1000명 이상인 데이타 필터링(추출)

df[(df['총이용인원']>=1000) & (df['선별']=='경부선')]
```
