```python
import numpy as np

df = pd.DataFrame(np.random.rand(100, 100)) # 0~1까지의 랜덤값을 100행 100열로...
df
```

```python
df = df[df>0.3] # 0.3보다 큰 데이터들을 다시 넣음(0.3보다 작은 값들은 결측치가 됨)
df
```

```python
import missingno as msno
msno.matrix(df) # 하얀색이 null값
```
