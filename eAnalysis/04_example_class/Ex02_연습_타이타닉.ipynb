```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
# 데이타 로딩 - 기존 데이타셋과 다르기에 확인 필요
raw_data = pd.read_csv('./data/titanic.csv')
raw_data.info()

raw_data.describe()
```

##  생존자와 사망자의 시각화

```python
raw_data.groupby('Survived').count()['PassengerId']
print(raw_data.groupby('Survived').count()['PassengerId'])
plt.bar(['사망자', '생존자'] ,raw_data.groupby('Survived').count()['PassengerId'])
```
