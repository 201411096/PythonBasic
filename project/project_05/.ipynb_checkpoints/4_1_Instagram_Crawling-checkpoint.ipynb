### 인스타그램 검색 결과 URL을 만들어서 접속하기

```python
#인스타그램 검색결과 URL을 만드는 함수
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/" + word
    return url   
```

```python
# selenium으로 URL 접속하기
from selenium import webdriver

# webdriver 설치 경로 맞추기
driver = webdriver.Chrome("D:/pythonwork/조별과제/5조 인스타그램 탑10지도/webdriver/chromedriver.exe")

# 검색어 변수에 담기
word = "성수동카페거리"
# URL 만드는 함수 연결
url = insta_searching(word)
# URL 주소로 띄우기
driver.get(url)
```

### 첫 번째 게시글 열기 

```python
# HTML에서 첫번째 게시글 찾아 클릭하기
def select_first(driver):
    # 로딩 속도 때문에 클릭이벤트를 못하기에 추가 ( 상황에 따라 실행되었다 안되었다 함 )
    driver.implicitly_wait(10) 
    # 태그 class 이름으로 찾아서 클릭버튼 연결
    first = driver.find_element_by_css_selector("div._9AhH0")
    first.click()
    time.sleep(3)
```

### 게시글 정보 가져오기

```python
# 게시글 정보 가져오기
import re

def get_content(driver):
    # ① 현재 페이지 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    # ② 본문 내용 가져오기
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content = ' '
    # ③ 본문 내용에서 해시태그 가져오기(정규식 활용)
    tags = re.findall(r'#[^\s#,\\]+', content)  
    # ④ 작성일자 정보 가져오기
    date = soup.select('time._1o9PC.Nzb55')[0]['datetime'][:10]
    # ⑤ 좋아요 수 가져오기
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1]   
    except:
        like = 0
    # ⑥ 위치정보 가져오기
    try: 
        place = soup.select('div.M30cS')[0].text
    except:
        place = ''
    # ⑦ 수집한 정보 저장하기
    data = [content, date, like, place, tags]
    return data
```

### 다음 게시글 열기

```python
# 다음 게시글 열기
def move_next(driver):
    # 로딩 속도 때문에 클릭이벤트를 못하기에 추가 ( 상황에 따라 실행되었다 안되었다 함 )
    driver.implicitly_wait(10) 
    # 태그 class 이름으로 찾아서 클릭버튼 연결
    right = driver.find_element_by_css_selector ('a.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(3)
```

### 로그인하기

```python
from selenium.webdriver.common.keys import Keys
# 인스타그램 아이디 비밀번호 변수에 담기
id="01077608253"
pwd="a91034658!"

# 로그인
def login(driver):
    driver.implicitly_wait(3)
    # 아이디와 비밀번호 태그 name으로 찾기
    driver.find_element_by_name('username').send_keys(id)
    elem=driver.find_element_by_name('password')
    elem.send_keys(pwd)
    elem.send_keys(Keys.RETURN)
    # 로그인 버튼 클릭하기
    right = driver.find_element_by_css_selector ('body > div.RnEpo._Yhr4 > div.pbNvD.fPMEg._8PWHW > div > div.Igw0E.IwRSH.eGOV_._4EzTm.MGdpg.lDRO1 > div > div > div.gr27e.o7laV > div > form > div:nth-child(4)')
    right.click()
    # 로그인 다음에 나오는 화면 넘어가기위해 button 클릭
    click=driver.find_element_by_css_selector ('section > main > div > div > div > div > button')
    click.click()
    
    
```

### 여러 게시글 정보 수집하기

```python
# 인스타그램 크롤링
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re

# ① 크롬 브라우저 열기
driver = webdriver.Chrome('D:/pythonwork/조별과제/5조 인스타그램 탑10지도/webdriver/chromedriver.exe') 

# ② 인스타그램 검색페이지 URL 만들기
word = "성수동카페거리"    #검색어
url = insta_searching(word)

# ③ 검색페이지 접속하기
driver.get(url)
time.sleep(3)

# ④ 첫 번째 게시글 열기
select_first(driver)

# 로그인 하기
login(driver)
select_first(driver)
# ⑤ 비어있는 변수(results)만들기
results = [ ]


# ⑥→⑦→⑧ 여러 게시물 수집하기
target = 500      # 크롤링할 게시글 수
for i in range(target):
    data = get_content(driver)    # 게시글 정보 가져오기
    results.append(data)
    move_next(driver)
    print(i)
    time.sleep(1) # 멈추는걸 방지해 1초 쉼.
```

```python
#크롤링 결과 저장하기

import pandas as pd

results_df = pd.DataFrame(results)
results_df.columns = ['content','data','like','place','tags']
results_df.to_excel('./files/홍대카페거리.xlsx') # 파일저장할 경로
results_df
```

###  여러 엑셀 파일의 중복을 제거한 후 통합 저장

인스타그램에서 여러 개의 키워드로 검색하거나 어러 번에 걸쳐 데이터를 수집하는 경우가 많다.

동일한 게시물을 중복으로 수집되는 경우 중복을 제거해야 한다.

```python
import pandas as pd
# 여러 개의 저장파일 통합하기
jeju_insta_df = pd.DataFrame( [ ] )

folder = './files/'
# 통합할 파일들 리스트에 담기
f_list = ['3_hayan_crawling_ssuMatJip.xlsx','4_1_crawling_result.xlsx','cafe.xlsx'
          ,'sungsudong.MatJip.xlsx','성수.xlsx','성수카페거리.xlsx']
# 반복문을 통해 통합.
for fname in f_list:
    fpath = folder + fname
    temp = pd.read_excel(fpath)
    sung_insta_df = sung_insta_df.append(temp)

    # 컬럼 지정.
sung_insta_df.columns =['index','content','data','like','place','tags']
sung_insta_df.shape
```

```python
# 중복 데이터 제거하고 저장하기(drop_duplicates)
sung_insta_df.drop_duplicates(subset = [ "content"] , inplace = True) # 내용이 중복되는것 드롭.
# 통합 + 중복제거된 파일 저장.
sung_insta_df.to_excel('./files/성수추천.xlsx',index = False)
sung_insta_df
```
